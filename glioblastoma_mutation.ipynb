```python
data= pd.read_excel("table_s14_bbac451.xlsx")
data.head(10)
```

we will organize the xcel file structure , to be normal again.

```python
data_cleaned = data.dropna(how='all')  # Drops rows that are completely NaN

# Step 2: Set proper header (row 1 is the actual header)

data_cleaned.columns = data_cleaned.iloc[0]
data_cleaned= data_cleaned[1:]
# Supprimer les colonnes dont le nom est NaN
data_cleaned = data_cleaned.loc[:, data_cleaned.columns.notna()]
data_cleaned = data_cleaned.reset_index(drop=True)

```

```python
data_cleaned.head(10)
print(data_cleaned.shape)
print(data_cleaned["Gene"])
print(data_cleaned.columns)
```

```python
#data_cleaned.iloc[:, 4:8]
data_cleaned.iloc[:, 11:14]
```

```python


# Step 5: Flatten the data (reshape) - we need to collapse the 4-column groups into one set of rows
data_flattened = pd.DataFrame()


# Concatenate the columns for UniProt, Gene, Mutation, Class for each group


data_flattened = pd.concat([
    data_flattened,
    data_cleaned.iloc[:, 0:4],
    data_cleaned.iloc[:, 4:8],
    data_cleaned.iloc[:, 8:12],
    data_cleaned.iloc[:, 12:16]
], axis=0)

# Reset the index for the flattened data
data_flattened = data_flattened.reset_index(drop=True)
data_flattened=data_flattened.dropna(how="all")



```
